# Split X_train and y_train

```python
train_data = df.iloc[:,1:]
train_data.head()
```

```python
train_x = train_data.iloc[:,:-1]
train_x.head()
```

```python
train_y = train_data.iloc[:,-1]
train_y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(train_x, train_y, test_size=0.25, random_state=123)
```

# Train Model

```python
#max_features
#n_estimators
#max_depth
#min_sample_split
#max_leaf_nodes
```

```python
rf = RandomForestClassifier(max_features = 500,n_estimators = 600,max_depth = 3,min_samples_split = 2,max_leaf_nodes = 30, random_state=123)
rf.fit(X_train, y_train)
```

# Prediction

```python
y_pred = rf.predict(X_test)
```

# Accuracy Evaluation

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Mean accuracy score: {accuracy:.3}')
```

# F1 Score Evaluation

```python
from sklearn.metrics import f1_score

f1 = f1_score(y_test, y_pred, average='macro')
print('f1 score :', f1)
```
