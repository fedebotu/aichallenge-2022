```python
def correlation_importance_plot(df, key, number=15, title='Correlation Importance'):
    correlations = df.corr()
    df_corr_cols = correlations[key]
    corr_ouput = df_corr_cols.sort_values(ascending=False, key=abs).dropna().drop(key) 
    # plot_data = corr_ouput[abs(corr_ouput)>threshold]
    plot_data = corr_ouput[:number]

    fig, ax = plt.subplots(1, 1, figsize=(12, 4))
    x = plot_data.index
    y = abs(plot_data.values)
    negatives = plot_data.values < 0

    pos_vals = [v > 0 for v in plot_data.values]
    neg_vals = [v < 0 for v in plot_data.values]
    max_pos_vals = float(max(pos_vals))
    min_neg_vals = float(min(neg_vals)) 

    # colors = cm.Reds(y / float(max(y)))
    colors = []
    for val in plot_data.values:
        if val > 0: colors.append('tab:red')
        else: colors.append('tab:blue')

    # print(colors)
    ax.bar(x,y, color=colors)
    plt.xticks(rotation='vertical')
    for p, neg in zip(ax.patches, negatives):
        if neg: mul = -1
        else: mul = 1
        ax.annotate(str(mul*round(p.get_height(),3)), (p.get_x() * 1.005, p.get_height() * 1.005))

    ax.set_ylabel('Absolute value of correlation with {}'.format(key))
    ax.set_title(title, fontweight='bold')
    # plt.savefig('Correlation Importance.pdf', bbox_inches='tight')
    plt.show()
    return plot_data
```

```python
correlation_importance_plot(drop_dup_df, 'leaktype_num')
```
